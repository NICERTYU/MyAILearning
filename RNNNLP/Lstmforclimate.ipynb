```python
import os


data_position = "../Dataset/"
absolute_path = os.path.abspath(data_position)
lines = []
header = []
fname = os.path.join(absolute_path, "jena_climate_2009_2016.csv")


with open(fname) as f:
	data = f.read()
      
	  
	lines = data.split('\n')
	header = lines[0].split(',')
	lines = lines[1:]


print(header)
print(len(lines))


```

```python
import numpy as np

float_data = np.zeros((len(lines), len(header) - 1))
for i, line in enumerate(lines):
    values = line.split(',')
    for j, value in enumerate(values[1:]):
        float_data[i][j] = float(value)

print(float_data)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd


temp = float_data[:, 1]
plt.plot(range(len(temp)), temp)
plt.show()
```
